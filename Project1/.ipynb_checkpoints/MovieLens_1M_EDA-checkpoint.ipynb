# EDA FOR MOVIELENS DATA
Contains:
1) Data for **6040 users**: user_id, age, gender,  occupation, zip_code
2) Data for **3883 movies**: movie_id, title, release_date, video_release_date, IMDb_URL and genre(one hot)
3) Data for **1,000,209 ratings**:  user_id , movie_id, rating,  timestamp

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import normalize
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```python
# 1M MovieLens
import seaborn as sns

pd.set_option("display.max_columns", None)
pd.set_option("display.width", None)

# -----Read------
# Ratings
ratings_1M = pd.read_csv(
    "datasets/movielens-1m/ratings.dat", 
    sep="::", 
    engine="python",  # required for multi-character separator
    names=["user_id", "movie_id", "rating", "timestamp"],
    encoding="latin-1"
)

# Users
users_1M = pd.read_csv(
    "datasets/movielens-1m/users.dat", 
    sep="::", 
    engine="python",
    names=["user_id", "gender", "age", "occupation", "zip_code"],
    encoding="latin-1"
)


# Movies
movies_1M = pd.read_csv(
    "datasets/movielens-1m/movies.dat", 
    sep="::", 
    engine="python",
    names=["movie_id", "title", "genres"],
    encoding="latin-1"
)
```

```python
users_1M.head()
```

```python
# Mapping age_bins and occupations

## age mapping
age_map = {
    1: "Under 18",
    18: "18-24",
    25: "25-34",
    35: "35-44",
    45: "45-49",
    50: "50-55",
    56: "56+"
}

# Create new column
users_1M["age_bin"] = users_1M["age"].map(age_map)

# Quick check
print(users_1M["age_bin"].unique())


## occupation mapping
occupation_map = {
    0: "other or not specified",
    1: "academic/educator",
    2: "artist",
    3: "clerical/admin",
    4: "college/grad student",
    5: "customer service",
    6: "doctor/health care",
    7: "executive/managerial",
    8: "farmer",
    9: "homemaker",
    10: "K-12 student",
    11: "lawyer",
    12: "programmer",
    13: "retired",
    14: "sales/marketing",
    15: "scientist",
    16: "self-employed",
    17: "technician/engineer",
    18: "tradesman/craftsman",
    19: "unemployed",
    20: "writer"
}

users_1M["occupation"] = users_1M["occupation"].map(occupation_map)

users_1M.head()
```

```python

#one hot the genres

print(movies_1M['genres'].head())
print(type(movies_1M['genres'].iloc[0]))


movies_1M['genres'] = movies_1M['genres'].str.split('|')


all_genres = sorted(set(g for sublist in movies_1M['genres'] for g in sublist))


for g in all_genres:
    movies_1M[g] = movies_1M['genres'].apply(lambda x: int(g in x))

movies_1M.head()

```

```python
# merge and check
# merge users characteristics, movie info, ratings together
data_1M = ratings_1M.merge(movies_1M, on='movie_id').merge(users_1M, on='user_id')
# drop useless columns
data_1M_cleaned = data_1M.drop(['timestamp', 'zip_code'], axis=1)
#check dropping
data_1M_cleaned.info()
data_1M_cleaned.head()
```

```python
# checking data counts
ratings_1M.info()
users_1M.info()
movies_1M.info()
```

## Distribution of ratings


```python


plt.hist(ratings_1M["rating"], bins=5, edgecolor="black", rwidth=0.8)
plt.title("Distribution of Movie Ratings")
plt.xlabel("Rating")
plt.ylabel("Count")
plt.xticks([1, 2, 3, 4, 5])
plt.show()
```

The ratings cluster in the mid-to-high range rather than being evenly spread across 15, possibly due to positivity bias(People are more likely to watch movies they expect to enjoy), selection bias(users who give ratings are usually enthusiasts, motivated to give ratings, skewing them upwards).

# EDA: Users


### Age Distribution

```python


age_order = ["Under 18", "18-24", "25-34", "35-44", "45-49", "50-55", "56+"]

sns.countplot(x="age_bin", data=users_1M, order=age_order)
plt.title("User Age Distribution")
plt.xticks(rotation=45)
plt.show()



```

Following the age bins provided by the MovieLens 1M README.txt, the most prominent age bin in this data is 25-34 years old. Thus, subsequent clustering on this data may result in overrepresentation for this age group.

### Gender Distribution

```python
sns.countplot(x="gender", data=users_1M)
plt.title("Gender Distribution")
plt.show()
```

There is a dominating gender which is the males. Similarly, this may result in clustering overrepresenting the interests of this gender.

### Occupation Distribution

```python
print(users_1M.columns)  
print(users_1M["occupation"].head())   # preview values
print(users_1M["occupation"].unique())
```

```python
# Occupation distribution
users_1M["occupation"].value_counts().plot(kind="bar", figsize=(12,4))
plt.title("Occupation Distribution")
plt.show()
```

The occupation distribution in the MovieLens 1M dataset is heavily skewed toward college/graduate students, followed by individuals in executive/managerial roles and those listed as other or not specified. By contrast, categories such as farmers, tradesmen/craftsmen, and unemployed users are minimally represented.

The demographic composition of the MovieLens 1M dataset shows notable imbalances across age, occupation, and gender. The most prominent age group is 2534 years old. Occupation data further reinforces this trend, with college/graduate students forming the largest group, alongside strong representation from executive/managerial users. At the same time, the dataset is disproportionately male, a pattern commonly noted in prior analyses of MovieLens. Taken together, these demographic skews suggest that the clusters and resulting recommendations are likely to be driven by the preferences of young, male students and professionals. As a result, certain user groups  such as older adults, women, or those in underrepresented occupations  may not have their preferences adequately captured, leading to potential biases in clustering outcomes and recommendation accuracy

```python
# each user has rated at least 20 movies 
user_rating_counts['num_ratings'].describe()
```

```python
# each movie has at least 1 rating
movie_rating_counts['num_ratings'].describe()
```

```python

# ratings per genre 
# *one movie can have multiple genres, not limited to 1
genre_cols = ratings_movies.columns[5:23]
genre_ratings_count = ratings_movies[genre_cols].sum().sort_values(ascending=False)

plt.figure(figsize=(10,6))
genre_ratings_count.plot(kind="bar")
plt.title("Number of Ratings per Genre")
plt.ylabel("Count of Ratings")
plt.xlabel("Genre")
plt.show()
```

# EDA: User-Movie Ratings 

```python
user_movies = ratings_movies.merge(users_1M, on='user_id', )
user_movies.info()
```

```python
# number of ratings per occupation
occupation_ratings = user_movies.groupby('occupation')['rating'].count().reset_index().rename(columns={"rating": "num_ratings"}).sort_values("num_ratings", ascending=False)
plt.figure(figsize=(12,6))
plt.bar(occupation_ratings["occupation"], occupation_ratings["num_ratings"])
plt.title("Number of Ratings by Occupation")
plt.xlabel("Occupation")
plt.ylabel("Number of Ratings")
plt.xticks(rotation=45, ha="right")  
plt.show()
```

```python
# number of ratings per age
age_ratings = user_movies.groupby('age_bin')['rating'].count().reset_index().rename(columns={"rating": "num_ratings"}).sort_values("num_ratings", ascending=False)
plt.figure(figsize=(12,6))
plt.bar(age_ratings["age_bin"], age_ratings["num_ratings"])
plt.title("Number of Ratings by Age")
plt.xlabel("Age")
plt.ylabel("Number of Ratings")
plt.show()
```

# Other EDA

```python
# looking into how high the average rating is for each 
# age group
avg_rating_age = user_movies.groupby("age_bin")["rating"].mean().reset_index()

plt.figure(figsize=(10,5))
sns.barplot(data=avg_rating_age, x="age_bin", y="rating", color="skyblue",  order=age_order)
plt.title("Average Rating by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Average Rating")
plt.xticks(rotation=45)
plt.show()

# Variance of ratings per age group
var_rating_age = user_movies.groupby("age_bin")["rating"].var().reset_index()

plt.figure(figsize=(10,5))
sns.barplot(data=var_rating_age, x="age_bin", y="rating", color="gold",  order=age_order)
plt.title("Variance of Ratings by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Variance")
plt.xticks(rotation=45)
plt.show()

```

```python
# for clearer representation of median can run this
plt.figure(figsize=(12,6))
sns.boxplot(x="age_bin", y="rating", data=user_movies,  order=age_order, 
            showmeans=True, 
            meanprops={"marker":"o", "markerfacecolor":"red", "markeredgecolor":"black"},
            medianprops={"color":"yellow", "linewidth":2})

plt.title("Distribution of Ratings by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Rating")
plt.show()
```

```python
# occupation group
avg_rating_occ = user_movies.groupby("occupation")["rating"].mean().reset_index()

plt.figure(figsize=(12,6))
sns.barplot(data=avg_rating_occ, x="occupation", y="rating", color="lightgreen")
plt.title("Average Rating by Occupation")
plt.ylabel("Average Rating")
plt.xticks(rotation=90)
plt.show()

# Variance of ratings per occupation
var_rating_occ = user_movies.groupby("occupation")["rating"].var().reset_index()

plt.figure(figsize=(10,5))
sns.barplot(data=var_rating_occ, x="occupation", y="rating", color="gold")
plt.title("Variance of Ratings by Occupation")
plt.xlabel("Occupation")
plt.ylabel("Variance")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(14,6))
sns.boxplot(data=user_movies, x="occupation", y="rating")
plt.title("Distribution of Ratings by Occupation")
plt.xlabel("Occupation")
plt.ylabel("Rating")
plt.xticks(rotation=45, ha="right")
plt.show()
```

```python

# gender group
avg_rating_gender = user_movies.groupby("gender")["rating"].mean().reset_index()

plt.figure(figsize=(6,4))
sns.barplot(data=avg_rating_gender, x="gender", y="rating", color="lightcoral")
plt.title("Average Rating by Gender")
plt.ylabel("Average Rating")
plt.show()
# Variance of ratings per gender

var_rating_gender = user_movies.groupby("gender")["rating"].var().reset_index()

plt.figure(figsize=(10,5))
sns.barplot(data=var_rating_gender, x="gender", y="rating", color="gold")
plt.title("Variance of Ratings by gender")
plt.xlabel("gender")
plt.ylabel("Variance")
plt.xticks(rotation=45)
plt.show()
```

When examining the distribution of ratings across demographic groups (age, occupation, and gender), average rating consistently hovered around 3.5, while the variance remained above 1.0 in nearly every subgroup. Suggests that users across different demographic categories tend to give broadly similar overall ratings, with no group showing a systematic tendency toward higher or lower ratings. At the same time, the relatively large within-group variance indicates substantial disagreement in preferences among individuals within the same demographic segment.
